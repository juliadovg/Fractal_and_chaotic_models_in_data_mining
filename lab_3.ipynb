```python
import numpy as np
import matplotlib.pyplot as plt

def fract_brownian_motion(n, h):
    """
        .
    
    :
    n (int):  
    h (float):  ,  [0, 1]
    
    :
    numpy.ndarray:    
    """
    t = np.linspace(0, 1, n)
    w = np.random.normal(0, 1, n)
    
    x = np.zeros_like(t)
    x[0] = w[0]
    
    for i in range(1, n):
        x[i] = x[i-1] + w[i] * ((t[i] - t[i-1])**(h))
    
    return x

#  
n = 512
h = 0.2
fbm = fract_brownian_motion(n, h)

#   
plt.figure(figsize=(12, 6))
plt.plot(fbm)
plt.title(f'    (H={h})')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()

#   
autocorr = np.correlate(fbm, fbm, 'full')
lags = np.arange(-len(fbm) + 1, len(fbm))

#    
plt.figure(figsize=(12, 6))
plt.plot(lags[len(fbm) - 1:], autocorr[len(fbm) - 1:])
plt.title(f'     (H={h})')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
