```python
import numpy as np
import matplotlib.pyplot as plt

def fract_brownian_motion(n, h):
    """
    Моделює одновимірний фрактальний броунівський рух.
    
    Аргументи:
    n (int): Кількість точок
    h (float): Показник Хьорста, діапазон [0, 1]
    
    Повертає:
    numpy.ndarray: Значення фрактального броунівського руху
    """
    t = np.linspace(0, 1, n)
    w = np.random.normal(0, 1, n)
    
    x = np.zeros_like(t)
    x[0] = w[0]
    
    for i in range(1, n):
        x[i] = x[i-1] + w[i] * ((t[i] - t[i-1])**(h))
    
    return x

# Згенерувати ФБР
n = 512
h = 0.2
fbm = fract_brownian_motion(n, h)

# Побудувати графік ФБР
plt.figure(figsize=(12, 6))
plt.plot(fbm)
plt.title(f'Одновимірний фрактальний броунівський рух (H={h})')
plt.xlabel('Час')
plt.ylabel('Значення')
plt.tight_layout()
plt.show()

# Обчислити автокореляційну функцію
autocorr = np.correlate(fbm, fbm, 'full')
lags = np.arange(-len(fbm) + 1, len(fbm))

# Побудувати графік автокореляційної функції
plt.figure(figsize=(12, 6))
plt.plot(lags[len(fbm) - 1:], autocorr[len(fbm) - 1:])
plt.title(f'Автокореляційна функція фрактального броунівського руху (H={h})')
plt.xlabel('Лаг')
plt.ylabel('Автокореляція')
plt.tight_layout()
plt.show()
```
