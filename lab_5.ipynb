```python
import numpy as np
import matplotlib.pyplot as plt

# Константи
p = 0.4  # Ймовірність виграшу в раунді
K = 100  # Початковий капітал
x_values = np.arange(1/512, 1 - 1/512 + 1/512, 1/512)  # Значення x від 1/512 до 1 - 1/512
B_values = K / x_values  # Бажаний капітал B як функція від x

# Функція для обчислення ймовірності P
def calculate_P(p, K, B):
    if p == 0.5:
        # Для спеціального випадку, коли p = 0.5
        return (B - K) / B
    else:
        q = 1 - p  # Ймовірність програшу
        # Формула для ймовірності досягнення мети
        return (1 - (q / p) ** K) / (1 - (q / p) ** B)

# Розрахунок значень P для кожного значення B
P_values = [calculate_P(p, K, B) for B in B_values]

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(x_values, P_values, label="P(x)", color="blue")
plt.title("Залежність ймовірності P від x")
plt.xlabel("x (K / B)")  # Вісь x
plt.ylabel("P")  # Вісь y
plt.grid(True)  # Відображення сітки
plt.legend()  # Легенда
plt.show()
```
